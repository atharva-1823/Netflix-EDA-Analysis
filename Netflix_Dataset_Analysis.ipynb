```python
import pandas as pd

# 1️⃣ Filter only Movies from original data
movies = data[data['type'] == 'Movie'].copy()

# 2️⃣ Create numeric duration column from "90 min" → 90.0
movies['new_dur'] = movies['duration'].str.extract(r'(\d+)').astype(float)

# 3️⃣ Handle missing country values (if any still there)
movies['country'] = movies['country'].fillna('International')

# 4️⃣ Split country string into list: "India, United States" → ["India", "United States"]
movies['country'] = movies['country'].str.split(', ')

# 5️⃣ Explode so each movie appears once per country
movies_exploded = movies.explode('country')

# 6️⃣ Group by country and calculate average duration
avg_dur = (
    movies_exploded
    .groupby('country')['new_dur']
    .mean()
    .sort_values(ascending=False)
)

# 7️⃣ (Optional) See top 10 countries by average movie duration
print(avg_dur.head(10))
```

```python
country_counts = movies_exploded['country'].value_counts()
```

```python
valid_countries = country_counts[country_counts >= 10].index
```

```python
filtered_movies = movies_exploded[movies_exploded['country'].isin(valid_countries)]
```

```python
avg_dur_clean = (
    filtered_movies.groupby('country')['new_dur']
    .mean()
    .sort_values(ascending=False)
)
```

```python
import matplotlib.pyplot as plt

# 1️⃣ Split data into Movies and TV Shows
movies = data[data['type'] == 'Movie']
shows  = data[data['type'] == 'TV Show']

# 2️⃣ Count titles per year (and sort by year)
movie_trend = movies['release_year'].value_counts().sort_index()
tv_trend    = shows['release_year'].value_counts().sort_index()

# 3️⃣ Put them into one DataFrame (aligns years automatically)
trend_df = (
    pd.concat([movie_trend, tv_trend], axis=1)
      .fillna(0)                # years that exist only in one series → 0
)
trend_df.columns = ['Movies', 'TV Shows']

# 4️⃣ Plot the trends
plt.figure(figsize=(12,6))
plt.plot(trend_df.index, trend_df['Movies'], marker='o', label='Movies')
plt.plot(trend_df.index, trend_df['TV Shows'], marker='o', label='TV Shows')

plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Netflix Release Trend: Movies vs TV Shows")
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

In this project, I performed an end-to-end Exploratory Data Analysis (EDA) of the Netflix dataset using Pandas for data processing and Matplotlib for visualization. The goal was to identify content trends, audience preferences, and Netflix’s platform growth over recent years.

⭐ Key Findings
1. Movies were added more frequently than TV Shows, especially after 2016, indicating Netflix’s aggressive expansion strategy.
2. The United States contributed the highest number of titles, followed by India, highlighting major content markets for Netflix.
3. After extracting numeric durations, India showed one of the highest average movie durations, reflecting long-form storytelling trends.
4. TV-MA and TV-14 are the most dominant ratings, showing that Netflix primarily focuses on mature and young-adult audiences.
5. International Movies, Dramas, and Comedies were among the most common genres, indicating broad viewer interest in global and story-driven content.
6. Netflix experienced rapid content growth from 2017 to 2020, with 2019 being the peak year of content addition.
7. Monthly analysis showed that July and December had the highest number of new titles, aligning with peak viewing periods like holidays and vacations.

⭐ Overall Conclusion
This analysis provides a clear picture of Netflix’s content strategy, including audience targeting, genre dominance, and growth patterns.
